```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("Big_Dance_CSV.csv")
df2 = pd.read_csv("kenpom barttorvik 2.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
for i in df1.index:
    for j in df2.index:
        if df1['Year'][i] == df2['YEAR'][j] and df1['Team1'][i] == df2['TEAM'][j]:
            arr1 = df2.iloc[j, 10:]
            break
            
    for k in df2.index:
        if df1['Year'][i] == df2['YEAR'][k] and df1['Team2'][i] == df2['TEAM'][k]:
            arr2 = df2.iloc[k, 10:]
            
    if i == 0:
        df3 = pd.DataFrame(np.subtract(arr1, arr2))
    else:
        df3[i] = np.subtract(arr1, arr2)

data = df3.transpose()
data.head()
```

```python
r = []

for i in df1.index:
    if df1['Score1'][i] > df1['Score2'][i]:
        r.append(1)
    else:
        r.append(0)
        
df1['Result'] = pd.Series(r, index=df1.index)
```

```python
seeddiff = []

for i in df1.index:
    seeddiff.append(df1['Seed1'][i] - df1['Seed2'][i])
        
df1['Seed_Diff'] = pd.Series(seeddiff, index=df1.index)
```

```python
df1.head()
```

```python
df1 = df1.drop(['Year','Region Number', 'Region Name', 'Score1', 'Score2'], axis=1)
df1.head()
```

```python
final = pd.concat([df1, data], axis=1)
final.head()
```

```python
final.to_csv('FULL TRAINING SET updated.csv', encoding='utf-8', index=False)
```
