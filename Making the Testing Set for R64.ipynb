```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("2024 round 1.csv")
df2 = pd.read_csv("kenpom barttorvik 2024.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
for i in df1.index:
    for j in df2.index:
        if df1['Team1'][i] == df2['TEAM'][j]:
            arr1 = df2.iloc[j, 10:]
            break
            
    for k in df2.index:    
        if df1['Team2'][i] == df2['TEAM'][k]:
            arr2 = df2.iloc[k, 10:]
            break
            
    if i == 0:
        df3 = pd.DataFrame(np.subtract(arr1, arr2))
    else:
        df3[i] = np.subtract(arr1, arr2)

data = df3.transpose()
data.head()
```

```python
seeddiff = []

for i in df1.index:
    seeddiff.append(df1['Seed1'][i] - df1['Seed2'][i])
        
df1['Seed_Diff'] = pd.Series(seeddiff, index=df1.index)
```

```python
final = pd.concat([df1, data], axis=1)
```

```python
final.head()
```

```python
final = final.drop(['Year','Region Number', 'Region Name', 'Seed1', 'Seed2'], axis=1)
final.head()
```

```python
final.to_csv('R64 TESTING SET.csv', encoding='utf-8', index=False)
```
